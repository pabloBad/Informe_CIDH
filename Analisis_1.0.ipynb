#### Crear función de transformación

```python
def process_name(name):
    # transforma un nombre al formato [primer_nombre, n_nombres, primer_apellido, segundo_apellido].
    # toma una lista de los 200 apellidos más comunes en chile.
    # solo procesa los nombres que estén en el listado de apellidos y que ademas, tengan por lo menos, 
    # 2 apellidos válidos al final. el resto debe hacerse a mano.
    
    if pd.isna(name) or pd.isnull(name):
        return [None, None, None, None]

    # recortar nombre. se transforman a minúsculas y se elimina cualquier carácter con tilde u otros símbolos.
    splitted_name = list(
        filter(lambda x: x != '',
               unidecode.unidecode(name).lower().split(' ')))
    
    # recorte sin modificación. se usa para retornar los datos.
    splitted_name_original_str = list(map(lambda x: x.capitalize(),
        filter(lambda x: x != '', name.split(' '))))

    # si hay por lo menos 3 strings en los nombres recortados, verificamos si los últimos 2 son apellidos.
    if (len(splitted_name) >= 3):
        
        detected_last_names = []        
        for string in splitted_name:
            try:
                detected_last_names.append(last_names_list.index(string))
            except:
                detected_last_names.append(None)

        # si los últimos 2 son apellidos, retorno (según su largo)
        #if not detected_last_names[-1] is None and not detected_last_names[-2] is None:
            
        # versión relajada: solo el primer apellido está en la lista:
        if not detected_last_names[-2] is None:
            
            # nombre, primer apellido, segundo apellido
            if len(splitted_name) == 3:
                return [
                    splitted_name_original_str[0], 
                    None,
                    splitted_name_original_str[1],
                    splitted_name_original_str[2]
                ]
            
            # primer nombre, segundo nombre, primer apellido, segundo apellido.
            elif len(splitted_name) == 4:
                return [
                    splitted_name_original_str[0],
                    splitted_name_original_str[1],
                    splitted_name_original_str[2],
                    splitted_name_original_str[3]
                ]
            
            # primer nombre, n nombres..., primer apellido, segundo apellido.
            else:
                return [
                    splitted_name_original_str[0],
                    ' '.join(splitted_name_original_str[1:-2]),
                    splitted_name_original_str[-2],
                    splitted_name_original_str[-1]
                ]
        else: 
            return [None, None, None, None]

    else:
        return [None, None, None, None]
```

#### Funcion para normalizar RUNs

La idea es que hagamos funciones para normalizar solo 1 caso y despues pasarlas por todo el dataset.
Así separas la lógica de procesamiento de un caso y como vas a iterar el dataset y como vas a guardar los datos.

```python
from itertools import cycle

def calc_dv(rut):
    reversed_digits = map(int, reversed(str(rut)))
    factors = cycle(range(2, 8))
    s = sum(d * f for d, f in zip(reversed_digits, factors))
    return (-s) % 11
```

```python
def normalize_RUN(run: str) -> list:
    '''
    Recibe solo 1 run y opcionalmente, su dígito verificador en el mismo string. 
    Transforma ese string en un arreglo que contendrá el rut válido mas su dígito verificador.
    '''

    # No tiene rut. retornar none
    if pd.isna(run) or pd.isnull(run):
        return [None, None]

    # Castear a str en caso que sea otro tipo
    if not (isinstance(run, str)):
        try:
            run = str(run)
        except:
            print('No pude castear el run {} a str'.format(run))
            
            
    # Dividir rut en valor y dígito verificador    
    splitted_run = run.split('-')
    
    # rut tiene dígito verificador.
    if len(splitted_run) == 2:
        splitted_run[0] = ''.join(splitted_run[0].split('.'))
        splitted_run[0] = ''.join(splitted_run[0].split(' '))
        splitted_run[0] = ''.join(splitted_run[0].split(','))
        splitted_run[0] = splitted_run[0].strip()
        
        splitted_run[1] = splitted_run[1].strip()
        splitted_run[1] = ''.join(splitted_run[1].split(','))
        splitted_run[1] = splitted_run[1].upper()

        # verificar dv. si es distinto al dado, entonces hay algo malo. retorno null.
        try: 
            calculated_dv = calc_dv(int(splitted_run[0]))
            calculated_dv = 'K' if calculated_dv == 10 else str(calculated_dv)
            
            if (calculated_dv != splitted_run[1]):
                return [None, None]
            # todo correcto. retorno el rut procesado y su dígito verificador.
            else:
                return splitted_run
            
        except Exception as e:
            print("Exception 1 during processing {}: {}".format(splitted_run, e))
            return [None, None]
        
    # Run tenia mas de un guion
    # Se asume ultimo guion es correcto
    elif len(splitted_run) > 2:
        
        # Obtenemos el dv
        dv = splitted_run[-1].upper()
        del splitted_run[-1]
        
        # juntamos el run y reposicionamos el dv
        splitted_run[0] = ''.join(splitted_run)
        splitted_run[1] = dv
        
        # Se continua de la misma manera que antes
        splitted_run[1] = splitted_run[1].strip()
        splitted_run[1] = ''.join(splitted_run[1].split(','))
        
        
        splitted_run[0] = ''.join(splitted_run[0].split('.'))
        splitted_run[0] = ''.join(splitted_run[0].split(' '))
        splitted_run[0] = ''.join(splitted_run[0].split(','))
        splitted_run[0] = splitted_run[0].strip()
        
        # verificar dv. si es distinto al dado, entonces hay algo malo. retorno null.
        try: 
            calculated_dv = calc_dv(int(splitted_run[0]))
            calculated_dv = 'K' if calculated_dv == 10 else str(calculated_dv)
            
            if (calculated_dv != splitted_run[1]):
                return [None, None]
            # todo correcto. retorno el rut procesado y su dígito verificador.
            else:
                return splitted_run
            
        except Exception as e:
            print("Exception 1 during processing {}: {}".format(splitted_run, e))
            return [None, None]
        
    # no tiene dígito verificador SEPARADO con guion 
    else:
        # calcularlo
        try:
            run = ''.join(run.split('.'))
            run = ''.join(run.split(' '))
            
            
            # Vemos si tenia el dv, pero no esta separado por guion
            dv = calc_dv(int(run[:-1]))
            dv = str(dv) if dv != 10 else 'K'
            possible_dv = run[-1].upper()
            
            # Si lo tenia, y era correcto.
            if possible_dv == dv:
                return [run[:-1], dv]
                
            # Si no lo tenia, lo agregamos
            dv = calc_dv(int(run))
            return [str(run), str(dv) if dv != 10 else 'K']

        # si no se puede, retorno none.
        except Exception as e:
            print("Exception 2 during processing {}: {}".format(run, e))
            return [None, None]
```

```python
def process_keyword_to_complaint(keyword):

    if pd.isna(keyword) or pd.isnull(keyword):
        return [None]

    if not isinstance(keyword, str):
        return [None]

    complaints_detected = []

    keyword = unidecode.unidecode(keyword.lower()).replace('.', '')

    for key in mapeo_palabras_clave_denuncias_proc:
        for sentence in mapeo_palabras_clave_denuncias_proc[key]:
            if sentence in keyword:
                complaints_detected.append(key)

    complaints_detected = list(set(complaints_detected))

    if len(complaints_detected) == 0:
        return [None]

    return [
        ', '.join(complaints_detected)
            
    ]
```

#### Funcion para transformar Contacto

```python
def process_contact(contact: str) -> list:
    '''
    Recibe un string de contacto. Puede ser correo, telefono o ambos. 
    Ademas puede tener entre parentesis el nombre/relacion de la persona de contacto.
    
    Transforma ese string en un arreglo que contendrá correo, telefono y nombre persona de contacto.
    '''

    # No tiene rut. retornar none
    if pd.isna(contact) or pd.isnull(contact):
        return [None, None, None]

    # Castear a str en caso que sea otro tipo
    if not (isinstance(contact, str)):
        try:
            contact = str(contact)
        except:
            print('No pude castear el contacto {} a str'.format(contact))

    # We check for 'SIN CONTACTO'
    if contact.upper() == 'SIN CONTACTO':
        return [None, None, None]
    
    # We check for 'PENDIENTE'
    if contact.upper() == 'PENDIENTE':
        return [None, None, None]
    
    # Puede tener sólo el telefono, se asume que es de la victima
    try:
        # Eliminamos puntos, espacios y '+'
        phone = contact.strip()
        phone = phone.replace('+','')
        phone = phone.replace('.','')
        phone = phone.replace(' ', '')
        phone = phone.rstrip('.')
        
        # Cast int to check if phone is a number
        phone = int(phone)
        
        return [None, str(phone), 'Víctima']
        
    except Exception as e:
        contact = contact.strip()
        
        # check for separators aka. 'splitter'
        
        # has an annotation
        if ('(' or ')') in contact:
            return [None, None, None]
        
        # The following do not have annotations, therefore we assume that the contact is the victim
        elif '//' in contact:
            # we use given separator
            splitted_contact = contact.split('//')
            #print(splitted_contact)
            
            # has at least 2 contacts: email & phone, phone & phone, email & email, other (not that order necessarily)
            # we check the **ONLY** first combination
            if len(splitted_contact) >= 2:
                # First is phone
                phone = splitted_contact[0]
                phone = phone.strip()
                phone = phone.replace('+','')
                phone = phone.replace(' ', '')
                phone = phone.rstrip('.') 

                email = splitted_contact[1]
                email = email.strip()
                email = email.replace('+', '')
                email = email.replace(' ', '')
                email = email.rstrip('.')
                if '@' in email:
                    # we check if phone is a number
                    try:
                        phone = int(phone)
                        return [email, str(phone), 'Víctima']
                    except Exception as e_1:
                        print("Exception 2.1 during processing phone {}: {}".format(phone, e_1))
                        return [None, None, None]
                
                # First was email!
                elif '@' in phone:
                    tmp = email
                    email = phone
                    phone = tmp
                    # we check if phone is a number
                    try:
                        phone = int(phone)
                        return [email, str(phone), 'Víctima']
                    except Exception as e_1:
                        print("Exception 3.1 during processing phone {}: {}".format(phone, e_1))
                        return [None, None, None]
                    
              
            # has more than 2 contacts
            else:
                return [None, None, None]
        
        elif '/' in contact:
            # we use given separator
            splitted_contact = contact.split('/')
            #print(splitted_contact)
            
            # has 2 contacts: email & phone, phone & phone, email & email, other (not that order necessarily)
            # we check the **ONLY** first combination
            if len(splitted_contact) >= 2:
                # First is phone
                phone = splitted_contact[0]
                phone = phone.strip()
                phone = phone.replace('+','')
                phone = phone.replace(' ', '')
                phone = phone.rstrip('.') 

                email = splitted_contact[1]
                email = email.strip()
                email = email.replace('+', '')
                email = email.replace(' ', '')
                email = email.rstrip('.')
                if '@' in email:                    
                    # we check if phone is a number
                    try:
                        phone = int(phone)
                        return [email, str(phone), 'Víctima']
                    except Exception as e_2:
                        print("Exception 2.2 during processing phone {}: {}".format(phone, e_2))
                        return [None, None, None]
                
                # First was email!
                elif '@' in phone:
                    tmp = email
                    email = phone
                    phone = tmp
                    # we check if phone is a number
                    try:
                        phone = int(phone)
                        return [email, str(phone), 'Víctima']
                    except Exception as e_2:
                        print("Exception 3.2 during processing phone {}: {}".format(phone, e_2))
                        return [None, None, None]
        
        elif '\n' in contact:
            # we use given separator
            splitted_contact = contact.split('\n')
            #print(splitted_contact)
            
            # has 2 contacts: email & phone, phone & phone, email & email, other (not that order necessarily)
            # we check the **ONLY** first combination
            if len(splitted_contact) >= 2:
                # First is phone
                phone = splitted_contact[0]
                phone = phone.strip()
                phone = phone.replace('+','')
                phone = phone.replace(' ', '')
                phone = phone.rstrip('.') 

                email = splitted_contact[1]
                email = email.strip()
                email = email.replace('+', '')
                email = email.replace(' ', '')
                email = email.rstrip('.')
                if '@' in email:                    
                    # we check if phone is a number
                    try:
                        phone = int(phone)
                        return [email, str(phone), 'Víctima']
                    except Exception as e_2:
                        print("Exception 2.2 during processing phone {}: {}".format(phone, e_2))
                        return [None, None, None]
                
                # First was email!
                elif '@' in phone:
                    tmp = email
                    email = phone
                    phone = tmp
                    # we check if phone is a number
                    try:
                        phone = int(phone)
                        return [email, str(phone), 'Víctima']
                    except Exception as e_2:
                        print("Exception 3.2 during processing phone {}: {}".format(phone, e_2))
                        return [None, None, None]
        
        # check for space separator
        elif ' ' in contact:
            # we use given separator
            splitted_contact = contact.split(' ')
            #print(splitted_contact)
            
            # has 2 contacts: email & phone, phone & phone, email & email, other (not that order necessarily)
            # we check the **ONLY** first combination
            if len(splitted_contact) >= 2:
                # First is phone
                phone = splitted_contact[0]
                phone = phone.strip()
                phone = phone.replace('+','')
                phone = phone.replace(' ', '')
                phone = phone.rstrip('.') 

                email = splitted_contact[1]
                email = email.strip()
                email = email.replace('+', '')
                email = email.replace(' ', '')
                email = email.rstrip('.')
                if '@' in email:
                    # we check if phone is a number
                    try:
                        phone = int(phone)
                        return [email, str(phone), 'Víctima']
                    except Exception as e_3:
                        print("Exception 2.3 during processing phone {}: {}".format(phone, e_3))
                        return [None, None, None]
                
                # First was email!
                elif '@' in phone:
                    tmp = email
                    email = phone
                    phone = tmp
                    # we check if phone is a number
                    try:
                        phone = int(phone)
                        return [email, str(phone), 'Víctima']
                    except Exception as e_3:
                        print("Exception 3.3 during processing phone {}: {}".format(phone, e_3))
                        return [None, None, None]
         
        # No more separators?
        # Only has email
        elif '@' in contact:
            return [contact, None, None]
        
        
        # Phone
        print("Exception 1 during processing phone {}: {}".format(contact, e))
        return [None, None, None] 
    
    
    
    
```
