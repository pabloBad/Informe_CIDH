#### Crear función de transformación

```python
def process_name(name):
    # transforma un nombre al formato [primer_nombre, n_nombres, primer_apellido, segundo_apellido].
    # toma una lista de los 200 apellidos más comunes en chile.
    # solo procesa los nombres que estén en el listado de apellidos y que ademas, tengan por lo menos, 
    # 2 apellidos válidos al final. el resto debe hacerse a mano.
    
    if pd.isna(name) or pd.isnull(name):
        return [None, None, None, None]

    # recortar nombre. se transforman a minúsculas y se elimina cualquier carácter con tilde u otros símbolos.
    splitted_name = list(
        filter(lambda x: x != '',
               unidecode.unidecode(name).lower().split(' ')))
    
    # recorte sin modificación. se usa para retornar los datos.
    splitted_name_original_str = list(map(lambda x: x.capitalize(),
        filter(lambda x: x != '', name.split(' '))))

    # si hay por lo menos 3 strings en los nombres recortados, verificamos si los últimos 2 son apellidos.
    if (len(splitted_name) >= 3):
        
        detected_last_names = []        
        for string in splitted_name:
            try:
                detected_last_names.append(last_names_list.index(string))
            except:
                detected_last_names.append(None)

        # si los últimos 2 son apellidos, retorno (según su largo)
        #if not detected_last_names[-1] is None and not detected_last_names[-2] is None:
            
        # versión relajada: solo el primer apellido está en la lista:
        if not detected_last_names[-2] is None:
            
            # nombre, primer apellido, segundo apellido
            if len(splitted_name) == 3:
                return [
                    splitted_name_original_str[0], 
                    None,
                    splitted_name_original_str[1],
                    splitted_name_original_str[2]
                ]
            
            # primer nombre, segundo nombre, primer apellido, segundo apellido.
            elif len(splitted_name) == 4:
                return [
                    splitted_name_original_str[0],
                    splitted_name_original_str[1],
                    splitted_name_original_str[2],
                    splitted_name_original_str[3]
                ]
            
            # primer nombre, n nombres..., primer apellido, segundo apellido.
            else:
                return [
                    splitted_name_original_str[0],
                    ' '.join(splitted_name_original_str[1:-2]),
                    splitted_name_original_str[-2],
                    splitted_name_original_str[-1]
                ]
        else: 
            return [None, None, None, None]

    else:
        return [None, None, None, None]
```

#### Funcion para normalizar RUNs

La idea es que hagamos funciones para normalizar solo 1 caso y despues pasarlas por todo el dataset.
Así separas la lógica de procesamiento de un caso y como vas a iterar el dataset y como vas a guardar los datos.

```python
from itertools import cycle

def calc_dv(rut):
    reversed_digits = map(int, reversed(str(rut)))
    factors = cycle(range(2, 8))
    s = sum(d * f for d, f in zip(reversed_digits, factors))
    return (-s) % 11
```

```python
def normalize_RUN(run: str) -> list:
    '''
    Recibe solo 1 run y opcionalmente, su dígito verificador en el mismo string. 
    Transforma ese string en un arreglo que contendrá el rut válido mas su dígito verificador.
    '''

    # No tiene rut. retornar none
    if pd.isna(run) or pd.isnull(run):
        return [None, None]

    # Castear a str en caso que sea otro tipo
    if not (isinstance(run, str)):
        try:
            run = str(run)
        except:
            print('No pude castear el run {} a str'.format(run))

    # Dividir rut en valor y dígito verificador
    splitted_run = run.split('-')

    # rut tiene dígito verificador.
    if len(splitted_run) == 2:
        splitted_run[0] = ''.join(splitted_run[0].split('.'))
        splitted_run[0] = ''.join(splitted_run[0].split(' '))
        
        splitted_run[1] = splitted_run[1].upper()
        
        # verificar dv. si es distinto al dado, entonces hay algo malo. retorno null.
        try: 
            calculated_dv = calc_dv(int(splitted_run[0]))
            calculated_dv = 'K' if calculated_dv == 10 else str(calculated_dv)
            
            if (calculated_dv != splitted_run[1]):
                return [None, None]
            # todo correcto. retorno el rut procesado y su dígito verificador.
            else:
                return splitted_run
            
        except Exception as e:
            print("Exception 1 during processing {}: {}".format(splitted_run, e))
            return [None, None]
        
        
    # no tiene dígito verificador
    else:
        # calcularlo
        try:
            run = ''.join(run.split('.'))
            run = ''.join(run.split(' '))

            dv = calc_dv(int(run))
            return [str(run), str(dv) if dv != 10 else 'K']

        # si no se puede, retorno none.
        except Exception as e:
            print("Exception 2 during processing {}: {}".format(run, e))
            return [None, None]
```

```python
def normalizeRUN(df_RUN, verbose=0):
    '''
    Recibe un df que tenga las columnas 'RUN' y 'Dígito verificador'.
    
    
    La columna DV debe estar vacía y el RUN debe tener el guión.
    
    Guarda los DV correspondientes en una lista.
    Elimina los guiones.
    Le agrega puntos al RUN que no tenga.
    
    Devuelve un df con 'RUN' y 'Dígito verificador'.
    Los RUN tienen puntos.
    '''
    run_guiones = np.array(df_RUN.RUN)
    if verbose:
        print("Los run con guiones son:")
        print(run_guiones, '\n')

    dv = [x[-1].upper() for x in run_guiones]
    if verbose:
        print("Se obtienen los DV")
        print(run_guiones, dv, '\n')


    if verbose:
        print("Actualizando run, eliminamos guion")
    for idx, run in enumerate(run_guiones):
        run = run[:-2]
        run_guiones[idx] = run
        if verbose:
            print(idx, "RUN final: ", run)
    
    if verbose:
        print("Agregamos puntos a los RUNs")
    for idx, run in enumerate(run_guiones):
        temp = addDotsRUN(run)
        run_guiones[idx] = temp
    
    if verbose:
        print("Actulizamos las columnas del df")
    
    df_RUN.RUN = run_guiones
    df_RUN['Dígito verificador'] = dv
    
    return df_RUN
    
    
def addDotsRUN(run):
    '''
    Normaliza el formato de un RUN sin puntos a uno con puntos.
    
    Recibe un run sin guion ni DV.
    '''
    run = str(run)
    
    # Verificamos que no tenga puntos
    if '.' in run:
        return run
    
    if '-' in run:
        return run
    
    
    largo = len(run)

    cant_dots = largo//3
    
    temp = [run]
    for i in range(1, cant_dots + 1):
        temp.append(temp[-1][:(largo - 3 * i)] + '.' + temp[-1][(largo - 3 * i):])
        
    return temp[-1]
```

```python
def process_keyword_to_complaint(keyword):

    if pd.isna(keyword) or pd.isnull(keyword):
        return [None]

    if not isinstance(keyword, str):
        return [None]

    complaints_detected = []

    keyword = unidecode.unidecode(keyword.lower()).replace('.', '')

    for key in mapeo_palabras_clave_denuncias_proc:
        for sentence in mapeo_palabras_clave_denuncias_proc[key]:
            if sentence in keyword:
                complaints_detected.append(key)

    complaints_detected = list(set(complaints_detected))

    if len(complaints_detected) == 0:
        return [None]

    return [
        ', '.join(complaints_detected)
            
    ]
```
